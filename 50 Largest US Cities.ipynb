```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import requests
import pandas as pd
from config2 import api_key
from pandas_function import make_df
import time
```

```python
#Opening and reading a city list by population and cleaning the dataframe. This list will be used to create the list of cities we will be running our analysis
file_one = "Population_List.csv"
file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")
newDF = file_one_df[["NAME","STNAME","POPESTIMATE2018"]]
```

```python
# Data is a mix of cities and state. So we need to remove all State Level Information
state_list = newDF["STNAME"].unique()
state_list = newDF["STNAME"].unique()
cond = newDF['NAME'].isin(state_list) == True
newDF.drop(newDF[cond].index, inplace = True)
```

```python
#Function to Clean city data
def City_Cleaner(df,value):
    Bad_City_DF = df[df['NAME'].str.contains(value)]
    Bad_City_List = Bad_City_DF['NAME']
    cond = df['NAME'].isin(Bad_City_List) == True
    df.drop(df[cond].index, inplace = True)
```

```python
#All the variables that were removed from the data to remove duplicates and also to clean the data such that it can be inputed into the main function
City_Cleaner(newDF,"County")
City_Cleaner(newDF,"(pt.)")
City_Cleaner(newDF,"(balance)")
City_Cleaner(newDF,"metropolitan government")
City_Cleaner(newDF,"Balance of")
City_Cleaner(newDF,"Hempstead")
City_Cleaner(newDF,"Washington")
City_Cleaner(newDF,"Parish")
```

```python
# Sort data by population largest to smallest
final_DF = newDF.sort_values(by=['POPESTIMATE2018'], ascending=False)
```

```python
# Remove city duplicates
final_DF = final_DF.drop_duplicates(subset ="NAME", keep="first") 
```

```python
#Remove all terms such as city or town
final_DF = final_DF.head(50)

final_DF['NAME'] = final_DF['NAME'].str.replace(' city', '')
final_DF['NAME'] = final_DF['NAME'].str.replace(' town', '')
final_DF
```
